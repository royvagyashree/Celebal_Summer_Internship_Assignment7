```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.model_selection import train_test_split

fraud = data[data['Class'] == 1]
normal = data[data['Class'] == 0].sample(n=492, random_state=42)
balanced_data = pd.concat([fraud, normal]).sample(frac=1, random_state=42).reset_index(drop=True)
X = balanced_data.drop('Class', axis=1)
y = balanced_data['Class']  # ✅ Make sure labels come from the same balanced dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.3, random_state=42)

classifier = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree Classifier": DecisionTreeClassifier()
}
for name, clf in classifier.items():
    print(f"\n==== {name} ====")
    clf.fit(X_train, Y_train)
    y_pred = clf.predict(X_test)

    print(f"Accuracy: {accuracy_score(Y_test, y_pred):.4f}")
    print(f"Precision: {precision_score(Y_test, y_pred):.4f}")
    print(f"Recall: {recall_score(Y_test, y_pred):.4f}")
    print(f"F1 Score: {f1_score(Y_test, y_pred):.4f}")
```
